
# **Step 02: Import All the Required Libraries**

```python
from llama_index import SimpleDirectoryReader, VectorStoreIndex
from llama_index.llms.palm import PaLM
from IPython.display import Markdown, display
from llama_index import ServiceContext
from llama_index import StorageContext, load_index_from_storage
import os
```



# Step 03: Load the PDF Files and Extract the Data From the PDF Files

```python
!mkdir data
```

```python
documents = SimpleDirectoryReader("data").load_data()
```

```python
print(documents)
```

# Step 04: Split the Text into Small Chunks and Create Embeddings for Each of the Text Chunks


```python
from google.colab import userdata
os.environ['GOOGLE_API_KEY'] = userdata.get('Google_API_Key')
```

```python
llm = PaLM()
```

```python
# Configure Service Context
service_context = ServiceContext.from_defaults(llm=llm, chunk_size=800, chunk_overlap=20)
```

```python
index = VectorStoreIndex.from_documents(documents, service_context=service_context)
```

# Step 05: Storing and Loading the Index

```python
index.storage_context.persist()
```

# Step 06: Q/A

```python
query_engine = index.as_query_engine()
while True:
  query = input("You: ")
  response = query_engine.query(query)
  display(Markdown(f"Bot: {response}"))

```
